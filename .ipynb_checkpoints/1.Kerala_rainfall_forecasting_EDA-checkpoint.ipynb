```python
# Convert date column to datetime
df['date'] = pd.to_datetime(df['date'])
```

```python
# Drop missing values
df.dropna(inplace=True)
```

## 2.Exploratory Data Analysis

```python
# Create additional time-based features for a better analysis
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day_of_year'] = df['date'].dt.dayofyear
df['month_name'] = df['date'].dt.month_name()
```

```python
# Define seasons of the year
def get_season(month):
    if month in [6,7,8,9]:     # June to September - Monsoon
        return 'Monsoon'
    elif month in [10, 11]:    # October to November - Post-Monsoon
        return 'Post-Monsoon'
    elif month in [12, 1, 2]:  # December to February - Winter
        return 'Winter'
    else:                      # March to May - Pre-Monsoon
        return 'Pre-Monsoon'

df['season'] = df['month'].apply(get_season)
df.head(3) 
```

```python
# Unique states and their data availability

state_info = df.groupby(['state_code', 'state_name']).agg({
    'date': ['min', 'max', 'count']
}).round(2)

state_info.columns = ['Start_Date', 'End_Date', 'Total_Records']
state_info = state_info.sort_values('state_code')
print(f"Total States/UTs: {len(state_info)}")
state_info
```

```python
# rainfall statistics by season
seasonal = df.groupby('season')['actual'].agg([
    'count', 'mean', 'std', 'min', 'max'])
seasonal
```

### Visualisations

```python
# Total rainfall per year
df['year'] = df['date'].dt.year
yearly_rainfall = df.groupby('year')['actual'].sum()

plt.figure(figsize=(10,5))
plt.plot(yearly_rainfall.index, yearly_rainfall.values, marker='o',color='#2ea7a8')
plt.title("Total Rainfall per Year",color='#3b29a3',fontweight='bold')
plt.xlabel("Year")
plt.ylabel("Rainfall (mm)")
plt.show()
```

```python
# Monthly rainfall patterns over the years
monthly_rainfall = df.groupby(['year', 'month'])['actual'].mean().reset_index()

# Create pivot table for heatmap
# A heatmap expects matrix-style data:
# Rows = each year, Columns = each month, Cell values = rainfall for that year/month
rainfall_pivot = monthly_rainfall.pivot(index='year',columns='month',values='actual')

# plot
plt.figure(figsize=(12, 8))
sns.heatmap(rainfall_pivot, cmap='Blues', annot=True, fmt='.1f',cbar_kws={'label': 'Average Rainfall (mm)'})
plt.title('Monthly Rainfall Patterns (2009-2024)',color='#3b29a3',fontweight='bold')
plt.xlabel('Month')
plt.ylabel('Year')
plt.yticks(rotation=0)
plt.show()
```

```python
# Monsoon season analysis
monsoon_data = df[df['season']=='Monsoon']
yearly = monsoon_data.groupby('year')['actual'].sum().reset_index()

plt.figure(figsize=(10,5))
plt.plot(yearly['year'], yearly['actual'], marker='o',color='#2ea7a8')
plt.title("Total Monsoon Rainfall by Year",color='#3b29a3',fontweight='bold')
plt.xlabel("Year")
plt.ylabel("Rainfall (mm)")
plt.show()
```

```python
# top & least wettest states
state_rainfall = df.groupby('state_name')['actual'].sum().sort_values(ascending=False)
top_states = state_rainfall.head(10)
least_states = state_rainfall.tail(10)

fig, axes = plt.subplots(1, 2, figsize=(10,6))
# Top 10 states
sns.barplot(x=top_states.index, y=top_states.values, color='#2ea7a8', ax=axes[0])
axes[0].set_title("Top 10 States by Rainfall", color='#3b29a3')
axes[0].set_ylabel("Rainfall (mm)")
axes[0].tick_params(axis='x', rotation=90)
# Least 10 states
sns.barplot(x=least_states.index, y=least_states.values, color='orange', ax=axes[1])
axes[1].set_title("Least 10 States by Rainfall", color='#3b29a3')
axes[1].set_ylabel("Rainfall (mm)")
axes[1].tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```

```python
# actual and forecasted rainfall over the years
diff = df.groupby('year')[['actual','rfs','normal']].sum().reset_index()

plt.figure(figsize=(10,5))
plt.plot(diff['year'],diff['actual'], marker='o',label='Actual Rainfall', color='#2ea7a8')
plt.plot(diff['year'], diff['rfs'], marker='o', label='Forecasted Rainfall', color='#f28e2b')
plt.plot(diff['year'],diff['normal'], marker='o',label='normal Rainfall', color='green')
plt.title('Actual, normal vs Forecasted Rainfall Over the Years',color='#3b29a3',fontweight='bold')
plt.xlabel("Year")
plt.ylabel("Rainfall (mm)")
plt.legend()
plt.show()
```

```python
# flood analysis
# Define flood as deviation > 200%
flood_threshold = 200

# Count drought and flood by state
flood_analysis = df.groupby('state_name').agg({
    'deviation': [
        'count',
        lambda x: (x > flood_threshold).sum()
    ]
}).round(2)

flood_analysis.columns = ['Total_Records', 'Flood_Days']
flood_analysis['Flood_Percentage'] = (flood_analysis['Flood_Days'] / flood_analysis['Total_Records'] * 100).round(2)

# Top states flood frequency
print("\nTop 10 States by Flood Frequency (% of days):")
print(flood_analysis.nlargest(10, 'Flood_Percentage')[['Flood_Days', 'Flood_Percentage']].to_string())
```
